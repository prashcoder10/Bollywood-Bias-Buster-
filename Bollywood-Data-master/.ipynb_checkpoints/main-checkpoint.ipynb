```python
mentions_df['character'] = mentions_df['character'].astype(str).str.lower()
gender_map = dict(zip(mentions_df['character'], mentions_df['gender']))
characters = mentions_df['character'].unique()
```

```python
for _, row in coref_df.dropna(subset=['coref plot']).iterrows():
    movie = row['movie title'] if 'movie title' in coref_df.columns else row.get('movie', 'Unknown')
    plot = row['coref plot']
    
    count_male = 0
    count_female = 0
    
    for char in characters:
        if char in plot:
            gender = gender_map.get(char)
            if gender:
                mentions = plot.count(char)
                if gender.upper() == 'MALE':
                    count_male += mentions
                elif gender.upper() == 'FEMALE':
                    count_female += mentions
    movie_gender_mentions = []
    movie_gender_mentions.append({
        'movie': movie,
        'male_mentions': count_male,
        'female_mentions': count_female
    })
```

```python
mention_df = pd.DataFrame(movie_gender_mentions)

# Plot per movie
mention_df.set_index('movie')[['male_mentions', 'female_mentions']].plot(
    kind='bar',
    figsize=(12, 6),
    color=['blue', 'pink']
)
plt.title("Mentions in Coreferenced Plots by Gender Per Movie")
plt.xlabel("Movie")
plt.ylabel("Mention Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
```

```python
female_verb = pd.read_csv("C:\\Users\\prash\\docu3c\\Bollywood-Data-master\\wikipedia-data\\cleaned_female_verb.csv", on_bad_lines='skip', header=None)
male_verb = pd.read_csv("C:\\Users\\prash\\docu3c\\Bollywood-Data-master\\wikipedia-data\\cleaned_male_verb.csv", on_bad_lines='skip', header=None)
```

```python
female_counts = Counter([v.lower() for v in female_verb[1].dropna()])
male_counts = Counter([v.lower() for v in male_verb[1].dropna()])
```

```python
def categorize_verbs(verb_counts):
    assertive = sum(count for verb, count in verb_counts.items() if verb in assertive_verbs)
    passive = sum(count for verb, count in verb_counts.items() if verb in passive_verbs)
    return assertive, passive

f_assertive, f_passive = categorize_verbs(female_counts)
m_assertive, m_passive = categorize_verbs(male_counts)
```

```python
df = pd.DataFrame({
    'Assertive': [f_assertive, m_assertive],
    'Passive': [f_passive, m_passive]
}, index=['Female', 'Male'])

# Plot
df.plot(kind='bar', stacked=True, title='Assertive vs Passive Verbs by Gender', color=['steelblue', 'lightcoral'])
plt.ylabel('Verb Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import fitz  # PyMuPDF
import os
```

```python
def extract_pdf_to_txt(pdf_path, output_txt_path):
    doc = fitz.open(pdf_path)
    full_text = "\n".join([page.get_text() for page in doc])
    
    with open(output_txt_path, 'w', encoding='utf-8') as f:
        f.write(full_text)
```

```python
extract_pdf_to_txt("C:\\Users\\prash\\docu3c\\Bollywood-Data-master\\scripts-data\\Neerja.pdf", "C:\\Users\\prash\\docu3c\\Bollywood-Data-master\\scripts-text\\Neerja.txt")
```

```python
pdf_folder = "scripts-data"
txt_folder = "scripts-text"
os.makedirs(txt_folder, exist_ok=True)

for filename in os.listdir(pdf_folder):
    if filename.endswith(".pdf"):
        pdf_path = os.path.join(pdf_folder, filename)
        txt_path = os.path.join(txt_folder, filename.replace(".pdf", ".txt"))
        extract_pdf_to_txt(pdf_path, txt_path)
```

```python
import spacy
import pandas as pd
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def process_script_text(txt_path):
    with open(txt_path, 'r', encoding='utf-8') as f:
        text = f.read()
        
    doc = nlp(text)
    
    data = []
    for token in doc:
        data.append({
            "token": token.text,
            "lemma": token.lemma_,
            "pos": token.pos_,
            "dep": token.dep_,
            "head": token.head.text,
            "ent_type": token.ent_type_,
            "is_stop": token.is_stop
        })
    
    return pd.DataFrame(data)
```

```python
df = process_script_text("C:\\Users\\prash\\docu3c\\Bollywood-Data-master\\scripts-text\\Neerja.txt")
print(df.head())
```

```python
nlp_outputs = []

for txt_file in os.listdir("scripts-text"):
    if txt_file.endswith(".txt"):
        movie_name = txt_file.replace(".txt", "")
        path = os.path.join("scripts-text", txt_file)
        df = process_script_text(path)
        df["movie"] = movie_name
        nlp_outputs.append(df)

# Combine all movies
final_df = pd.concat(nlp_outputs, ignore_index=True)
final_df.to_csv("processed_scripts_nlp.csv", index=False)
```
